```python
df_delhi1=delhi[delhi['Place'].isin(df_delhi['Place'].values)]
# len(df_delhi1['Place'].unique())
print(df_delhi1)
df_mumbai1=mumbai[mumbai['Place'].isin(df_mumbai['Place'].values)]
print(df_mumbai1.shape)
```
